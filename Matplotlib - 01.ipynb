```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
plt.plot(x, y, "r--")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("X versus Y")
```

```python
plt.subplot(1, 2, 1) # 1 linha e 2 colunas. O terceiro parâmetro mostra com qual gráfico está trabalhando.
plt.plot(x, y, "r--")
plt.subplot(1, 2, 2)
plt.plot(x, y, "g*-")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Dimensoes do gráfico no espaço
axes1.set_xlabel("Eixo X")   
axes1.set_ylabel("Eixo Y")
axes1.set_title("Título")
axes1.plot(x, y, "b-*")

axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])
axes2.set_xlabel("Eixo X")
axes2.set_ylabel("Eixo Y")
axes2.set_title("Título")
axes2.plot(x, y, "r--")
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, "b--")
ax.plot(x, x**4, "g-.")
ax.set_xlabel("Eixo X")
ax.set_ylabel("Eixo Y")
ax.set_title("Título")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=5)
```

```python
for axe in ax:
    axe.plot(x, y, "b")
    axe.set_title("Título")
fig
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout() # Arruma melhor as sobreposições dos gráficos
```

```python
fig = plt.figure(figsize=(8, 4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(6, 6))

axes.plot(x, y, "r-", label="x^2")
axes.plot(y, x, "g-", label="x**1/2")
axes.set_title("Título")
axes.legend(loc=1) #Mostra a legenda feita no 'label'. O 'loc=' define a posição que ficará a legenda
```

```python
fig.savefig("imagem.jpg")
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(x, x**2, "g--", linewidth=2) #linewidth = espessura das linhas | alpha = opacidade das linhas
ax.plot(x, x**3, "b-*", linewidth=10, alpha=0.5)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(x, x**2, "g--", linewidth=2) #linewidth = espessura das linhas | alpha = opacidade das linhas
ax.plot(x, x**3, "b-*", linewidth=1, alpha=0.5)

ax.set_xlim([0, 2])
ax.set_ylim([0, 10])
```
