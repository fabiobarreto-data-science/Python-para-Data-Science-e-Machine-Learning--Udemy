```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
plt.plot(x, y, "r--")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("X versus Y")
```

```python
plt.subplot(1, 2, 1) # 1 linha e 2 colunas. O terceiro parâmetro mostra com qual gráfico está trabalhando.
plt.plot(x, y, "r--")
plt.subplot(1, 2, 2)
plt.plot(x, y, "g*-")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Dimensoes do gráfico no espaço
axes1.set_xlabel("Eixo X")
axes1.set_ylabel("Eixo Y")
axes1.set_title("Título")
axes1.plot(x, y, "b-*")

axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])
axes2.set_xlabel("Eixo X")
axes2.set_ylabel("Eixo Y")
axes2.set_title("Título")
axes2.plot(x, y, "r--")
```
