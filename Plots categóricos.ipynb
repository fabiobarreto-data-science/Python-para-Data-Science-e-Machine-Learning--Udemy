```python
import seaborn as sns
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.barplot(x="sex", y="total_bill", data=tips) # MÃ©dia com desvio padrÃ£o
```

```python
sns.barplot(x="sex", y="total_bill", data=tips, estimator=np.std) # Desvio padrÃ£o do desvio padrÃ£o? Ainda nÃ£o entendi
```

```python
sns.countplot(x="sex", data=tips) # Conta o nÃºmero de homens e mulheres
```

```python
sns.boxplot(x="day", y="total_bill", data=tips) # Calcula os perscentis 25%, 50%, 75% e 100%
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette="rainbow") # Para saber a diferenÃ§a entre ter e nÃ£o ter o 'palette'
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette="rainbow", hue="sex")
```

```python
sns.boxplot(data=tips, palette="rainbow", orient="v" ) # Diz a lenda que nÃ£o serve para muita coisa v= vertical e h=horizontal
```

```python
sns.violinplot(x="day", y="tip", data=tips, hue="sex")
```

```python
sns.violinplot(x="day", y="tip", data=tips, hue="sex", split=True) # Metade masculino com feminino
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue="sex", dodge=True) # dodge = divide em masculino e feminino
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex", dodge=True, size=4)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, dodge=True, size=4, color="black") # Podemos fazer sobreposiÃ§Ãµes de PLOTS
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
sns.catplot(x="sex", y="total_bill", data=tips, kind="bar") # O KIND assume um tipo de plot dado na aula.
```
