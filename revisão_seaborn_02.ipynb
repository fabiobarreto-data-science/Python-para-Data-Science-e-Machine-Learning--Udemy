```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.barplot(x=tips["sex"], y=tips["total_bill"])
```

```python
sns.barplot(x=tips["sex"], y=tips["total_bill"], estimator=np.std);
```

```python
sns.countplot(x="sex", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette="rainbow", hue="sex")
```

```python
sns.boxplot(x="total_bill", y="time", data=tips, palette="rainbow", hue="sex", orient="h");
```

```python
sns.violinplot(x="day", y="tip", data=tips, palette="rainbow", hue="sex");
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True, hue="sex", dodge=True);
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex");
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex");
sns.violinplot(x="day", y="total_bill", data=tips, palette="rainbow");
```

```python
sns.catplot(x="sex", y="total_bill", data=tips, kind="bar");
```

```python
sns.catplot(x="sex", y="total_bill", data=tips, kind="violin");
```
