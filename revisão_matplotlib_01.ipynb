```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x*x
```

```python
y
```

```python
# FORMA FUNCIONAL
plt.plot(x, y, "g-*");
plt.title("X versus Y")
plt.xlabel("Valores de X")
plt.ylabel("Valores de Y")
```

```python
# EXISTE O 'SUBPLOT' E O 'SUBPLOTS'
plt.subplot(1, 2, 1);
plt.plot(x, y, "r--");

plt.subplot(1, 2, 2);
plt.plot(y, x, "b--");
```

```python
fig = plt.figure();
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]);
axes2 = fig.add_axes([1.1, 0.1, 0.8, 0.8]);
axes1.plot(x, y);
axes2.plot(y, x);

axes1.set_title("Título do gráfico 1");
axes1.set_xlabel("Valores de X");
axes1.set_ylabel("Valores de Y");

axes2.set_title("Título do gráfico 2");
axes2.set_xlabel("Valores de X");
axes2.set_ylabel("Valores de Y");
```

```python
fig, ax = plt.subplots();

ax.plot(x, y, "ro-");
ax.plot(x, x**4, "g--");
ax.plot(x, x**5, "b-*");

ax.set_xlabel("Valores de X");
ax.set_ylabel("Valores de X na potência");
ax.set_title("Título");
```

```python
fig, ax = plt.subplots(1, 3);

ax[0].plot(x, y, "ro-");
ax[1].plot(x, x**4, "g--");
ax[2].plot(x, x**5, "b-*");
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 3));

ax[0].plot(x, y, "r--", label="x versus y");
ax[0].set_title("Gráfico 1");
ax[0].set_xlabel("Valores de X");
ax[0].set_ylabel("Valores de Y");
ax[0].legend(loc=2)   # loc => localização da legenda

ax[1].plot(y, x, "+-", label="y versus x");
ax[1].set_title("Gráfico 2");
ax[1].set_xlabel("Valores de Y");
ax[1].set_ylabel("Valores de X");
ax[1].legend(loc=4)    # loc => localização da legenda
```

```python
fig.savefig("imagem.jpg")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 8))

ax[0].plot(x, x**2, "r:", label="x versus x²", linewidth=9, alpha=0.5, marker="o", markersize=8)
ax[0].set_title("Título")
ax[0].set_xlabel("x")
ax[0].set_ylabel("x²")
ax[0].legend(loc=2)
ax[0].set_xlim([0, 2])

ax[1].plot(x, x**3, color="green", label="x versus x³", linewidth=5, alpha=.3, marker="*", markersize=15)
ax[1].set_title("Título")
ax[1].set_xlabel("x")
ax[1].set_ylabel("x³")
ax[1].legend(loc=2)
ax[1].set_ylim([0, 60])
```

```python
#SCATTER
plt.scatter(x, y, marker="*", lw=5, label="SCATTER");

plt.title("Título")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend(loc=4)
```
