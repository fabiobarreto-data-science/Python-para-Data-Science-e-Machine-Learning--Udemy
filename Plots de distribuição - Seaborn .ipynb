```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.displot(tips["total_bill"], kde=True, bins=50) # displot  Umas das funções para apresentar o gráfico. kde = linha que passa pelo gráfico
# bins = quantas partes posso dividir o gráfico, no exemplo dado foi 50
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex") # Faz uma relação entra as duas variáveis x e y
```

```python
sns.pairplot(tips, hue="day", palette="coolwarm") # Pega as variáveis numéricas e faz a jointplot em cada uma delas. hue = separa dados diferentes
```

```python
sns.rugplot(tips["total_bill"])
```

```python
sns.kdeplot(tips["total_bill"])
sns.rugplot(tips["total_bill"])
```
